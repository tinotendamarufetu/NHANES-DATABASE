```python
# Function to determine the data type based on the first row of data
def get_data_type(value):
    try:
        int(value)
        return "INT"
    except ValueError:
        try:
            float(value)
            return "FLOAT"
        except ValueError:
            return "VARCHAR(255)"
```

```python
# Function to generate the CREATE TABLE statement
def generate_create_table_statement(file_path):
    table_name = os.path.splitext(os.path.basename(file_path))[0]
    with open(file_path, 'r') as csv_file:
        reader = csv.reader(csv_file)
        header = next(reader)  # Get the header row
        data_types = [get_data_type(next(reader)[i]) for i in range(len(header))]  # Get the data types based on the first row of data
        columns = [f"{col} {data_type}" for col, data_type in zip(header, data_types)]
        create_table_statement = f"CREATE TABLE {table_name} ({', '.join(columns)});"
        return create_table_statement

# Specify the folder path where the CSV files are located
folder_path = "C:/Users/tinot/Downloads/NHANES/All Laboratory Data/Dioxins, Furans, & Coplanar PCBs"

# Create the output file
output_file_path = os.path.join(folder_path, "create_table_statements.txt")
with open(output_file_path, 'w') as output_file:
    # Iterate over all CSV files in the folder
    for file_name in os.listdir(folder_path):
        if file_name.endswith(".csv"):
            file_path = os.path.join(folder_path, file_name)
            create_table_statement = generate_create_table_statement(file_path)
            output_file.write(create_table_statement + "\n\n")
            print(f"Created CREATE TABLE statement for {file_name}")
```

#Loop through each folder and convert all .xpt files to csv

```python
import os
import csv
import re

# Function to determine the data type based on the first row of data
def get_data_type(value):
    try:
        int(value)
        return "INT"
    except ValueError:
        try:
            float(value)
            return "FLOAT"
        except ValueError:
            return "VARCHAR(255)"

# Function to generate the CREATE TABLE statement
def generate_create_table_statement(file_path):
    table_name = os.path.splitext(os.path.basename(file_path))[0]
    with open(file_path, 'r') as csv_file:
        reader = csv.reader(csv_file)
        header = next(reader)  # Get the header row
        data_types = [get_data_type(next(reader)[i]) for i in range(len(header))]  # Get the data types based on the first row of data
        columns = [f"{col} {data_type}" for col, data_type in zip(header, data_types)]
        create_table_statement = f"CREATE TABLE {table_name} ({', '.join(columns)});"
        return create_table_statement

# Set the parent directory path where the folders containing CSV files are located
parent_directory = "C:/Users/tinot/Downloads/NHANES/All Laboratory Data"

# Loop through all subdirectories in the parent directory
for subdir in os.listdir(parent_directory):
    subdir_path = os.path.join(parent_directory, subdir)
    
    # Check if the current item is a directory
    if os.path.isdir(subdir_path):
        # Create the output file for the current subdirectory
        output_file_path = os.path.join(subdir_path, "create_table_statements.txt")
        with open(output_file_path, 'w') as output_file:
            # Loop through all CSV files in the current subdirectory
            for file_name in os.listdir(subdir_path):
                if file_name.endswith(".csv"):
                    file_path = os.path.join(subdir_path, file_name)
                    create_table_statement = generate_create_table_statement(file_path)
                    output_file.write(create_table_statement + "\n\n")
                    print(f"Created CREATE TABLE statement for {file_name}")
```
