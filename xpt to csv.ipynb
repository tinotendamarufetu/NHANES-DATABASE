#This one converts items in a folder only

```python
import pandas as pd
import os
```

```python
# Set the directory path where the .xpt files are located
directory = "C:/Users/tinot/Downloads/NHANES/All Laboratory Data/Human epididymal secretory protein E4 (HE4) - Serum (Surplus)"
```

```python
# Loop through all files in the input directory
for filename in os.listdir(directory):
    # Check if the file is an .xpt file
    if filename.endswith(".XPT"):
        # Read the .xpt file using pandas
        xpt_file = os.path.join(directory, filename)
        df = pd.read_sas(xpt_file)

        # Convert the dataframe to a .csv file
        csv_filename = os.path.splitext(filename)[0] + ".csv"
        csv_file = os.path.join(directory, csv_filename)
        df.to_csv(csv_file, index=False)

        print(f"Converted {filename} to {csv_filename}")
```

#Loop through each folder and convert all .xpt files to csv

```python
import pandas as pd
import os

# Set the parent directory path where the folders containing .xpt files are located
parent_directory = "C:/Users/tinot/Downloads/NHANES/Laboratory"

# Loop through all subdirectories in the parent directory
for subdir in os.listdir(parent_directory):
    subdir_path = os.path.join(parent_directory, subdir)
    
    # Check if the current item is a directory
    if os.path.isdir(subdir_path):
        # Loop through all files in the current subdirectory
        for filename in os.listdir(subdir_path):
            # Check if the file is an .xpt file
            if filename.endswith(".XPT"):
                # Read the .xpt file using pandas
                xpt_file = os.path.join(subdir_path, filename)
                df = pd.read_sas(xpt_file)

                # Convert the dataframe to a .csv file
                csv_filename = os.path.splitext(filename)[0] + ".csv"
                csv_file = os.path.join(subdir_path, csv_filename)
                df.to_csv(csv_file, index=False)

                print(f"Converted {filename} to {csv_filename}")
```
